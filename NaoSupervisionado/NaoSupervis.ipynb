```python
# Criando dados
```

```python
from sklearn.datasets import make_blobs
```

```python
data, _ = make_blobs(n_samples=300,n_features=2,centers=3,cluster_std=1.5)
```

```python
data
```

```python
## Analise Exploratória
#-----------------------
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.dpi"]=100
plt.scatter(data[:,0],data[:,1], marker='o', edgecolors="black")
plt.show()
```

Modelo K-Means

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit(data)
```

```python
#Predições
```

```python
resposta = model.predict(data)
```

```python
grupo1 = data[resposta == 0]
grupo2 = data[resposta == 1]
grupo3 = data[resposta == 2]

centros =model.cluster_centers_
```

```python
plt.scatter(grupo1[:,0],grupo1[:,1],marker='o',c='blue',label= "Grupo 1", edgecolors="Black")
plt.scatter(grupo2[:,0],grupo2[:,1],marker='s',c='red',label= "Grupo 2", edgecolors="Black")
plt.scatter(grupo3[:,0],grupo3[:,1],marker='v',c='green',label= "Grupo 3", edgecolors="Black")

plt.scatter(centros[:,0],centros[:,1],marker='*',c='yellow',label= "Centros",s=100)

plt.legend(scatterpoints= True)
plt.show()
```

```python
# Métricas
```

```python
from sklearn.metrics import calinski_harabasz_score, silhouette_score

sh = silhouette_score(data,model.labels_)
ch = calinski_harabasz_score(data, model.labels_)
```

```python
print(f'SH: {sh}')
print(f'CH: {ch}')
```
